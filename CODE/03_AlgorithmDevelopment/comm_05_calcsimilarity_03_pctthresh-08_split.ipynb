## Notebook 05 - Calculate / Adjust Similarity Scores


We'll need to investigate scaling and adjusting the similarity scores

---

### Import libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import time
```

```python
def num_uniques(ser):
    try:
        return len(ser.unique())
    except:
        return "Not unique check-able"
```

```python
def summarize_df(df):
    print("======DATA SUMMARY======")
    print("{} rows by {} columns".format(df.shape[0], df.shape[1]))
    print("\n======COLUMNS======")
    print(df.dtypes)
    print("\n======PREVIEW======")
    display(df.head())
    print("\n======NUMERICAL COL SUMMARY======")
    print(df.describe())
    print("\n")
    for col in df.columns:
        print("{}: {} unique values".format(col, num_uniques(df[col])))
```

---
### Import data

```python
%%time
print(time.asctime(time.localtime(time.time())))
dist = pd.read_parquet('community_distances_pctthresh-08_split.parquet')
summarize_df(dist)
```

```python
%%time
print(time.asctime(time.localtime(time.time())))
sizes = pd.read_csv('node_sizes.csv')[['community_id', 'count']]
summarize_df(sizes)
```

```python
%%time
print(time.asctime(time.localtime(time.time())))
asgn = pd.read_parquet('community_assignments_pctthresh-08.parquet') \
            .rename({'recipe_id': 'id'}, axis=1)
summarize_df(asgn)
```

```python
# Invert the distance to get a similarity score
new = pd.DataFrame(np.array([dist['community_id'].unique(), dist['community_id'].unique()]).T,
                   columns=['community_id', 'community_id_2']) \
        .astype({'community_id':'int64', 'community_id_2':'int64'})
new = new.merge(sizes, on='community_id')
new['count'] = new['count']/new['count'].max()
new['community_distance'] = dist['community_distance'].min() - (dist['community_distance'].max() - dist['community_distance'].min())*.25
new = new.drop('count', axis=1)
dist = pd.concat([dist, new])
dist['similarity'] = dist['community_distance'].max() - dist['community_distance'] \
                                + dist['community_distance'].min()
print("Similarity")
print(dist['similarity'].describe())
display(dist['similarity'].plot.hist(bins=20))
plt.show()

# Add scores 5% above the max for same-cluster distances
dist['community_id'].unique()

# Scale score
dist['similarity_scaled'] = dist['similarity'].apply(lambda x: (x - dist['similarity'].min()) /  \
                                                     (dist['similarity'].max() - dist['similarity'].min()))
print("\nSimilarity Scaled")
print(dist['similarity_scaled'].describe())
display(dist['similarity_scaled'].plot.hist(bins=20))
plt.show()
```

### Remove cluster 29 (25 recipes, far from other clusters) and rerun

```python
dist_filter = dist[(dist['community_id'] != 29) & (dist['community_id_2'] != 29)]
```

```python
print("Similarity")
print(dist_filter['similarity'].describe())
display(dist_filter['similarity'].plot.hist(bins=30))
plt.show()

dist_filter['similarity_scaled'] = dist_filter['similarity'].apply(lambda x: (x - dist_filter['similarity'].min()) /  \
                                                     (dist_filter['similarity'].max() - dist_filter['similarity'].min()))

print("\nSimilarity Scaled")
print(dist_filter['similarity_scaled'].describe())
display(dist_filter['similarity_scaled'].plot.hist(bins=30))
plt.show()
```
