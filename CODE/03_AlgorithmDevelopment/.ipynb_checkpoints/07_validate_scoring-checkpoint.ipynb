### Import libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import time
import sys
import seaborn as sns
```

```python
def num_uniques(ser):
    try:
        return len(ser.unique())
    except:
        return "Not unique check-able"

def summarize_df(df):
    print("======DATA SUMMARY======")
    print("{} rows by {} columns".format(df.shape[0], df.shape[1]))
    print("\n======COLUMNS======")
    print(df.dtypes)
    print("\n======PREVIEW======")
    display(df.head())
    print("\n======NUMERICAL COL SUMMARY======")
    print(df.describe())
    print("\n")
    for col in df.columns:
        print("{}: {} unique values".format(col, num_uniques(df[col])))
```

```python
%%time
print(time.asctime(time.localtime(time.time())))
local_path = '../05_RecipeExplorationTool/data/community_distances_pctthresh-08_split_adj.parquet'
sim_scores = pd.read_parquet(local_path)
summarize_df(sim_scores)
```

```python
%%time
print(time.asctime(time.localtime(time.time())))
local_path = 'community_assignments_pctthresh-08_split_filled.parquet'
asgn = pd.read_parquet(file_root+local_path)
summarize_df(asgn)
```

```python
# Sample shared_users dataset for runtime reduction - temporary
shared_users = shared_users.sample(frac=.10)
```

```python
shared_users = shared_users.merge(asgn, on='recipe_id')
shared_users = shared_users.merge(asgn, left_on='recipe_id_2', right_on='recipe_id', suffixes=('', '_2'))
```

```python
shared_users = shared_users.merge(sim_scores, on=['community_id', 'community_id_2'])
```

```python
sizes = shared_users.groupby(['pct_users']).agg({'similarity_scaled':['mean','count', 'median']}).reset_index()
sizes.columns = ['_'.join(a) if len(a[1]) > 1 else a[0] for a in sizes.columns]
```

```python
fig, ax = plt.subplots()
sns.scatterplot(data=sizes, x='pct_users', y='similarity_scaled_mean', s=10, hue='similarity_scaled_count', ax=ax)
sns.regplot(data=sizes,  x='pct_users', y='similarity_scaled_mean', logx=True, scatter=False, ax=ax, 
            line_kws={'color':'black'})
```

```python
fig, ax = plt.subplots()
sns.scatterplot(data=sizes, x='pct_users', y='similarity_scaled_median', s=10, hue='similarity_scaled_count', ax=ax)
sns.regplot(data=sizes,  x='pct_users', y='similarity_scaled_median', logx=True, scatter=False, ax=ax, 
            line_kws={'color':'black'})
```
